```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def left_sum(func, a, b, n, kind="equal"):
    def f(x):
        return eval(func)

    if kind == "equal":
        width = abs((b - a)/n)
        area_sum = 0
        for i in range(n):
            x = a + i * width
            area_sum += f(x)

        x = np.linspace(a, b, n+1)
        y = f(x)
        plt.bar(x[:-1], y[:-1], width=width, align="edge", edgecolor="black")
        plt.plot(x[:-1], y[:-1], marker="", color="black", markersize = 10)
        plt.show();
        plt.clf()
        return area_sum * width
```

```python
def right_sum(func, a, b, n, kind="equal"):
    def f(x):
        return eval(func)

    if kind == "equal":
        width = abs((b - a)/n)
        area_sum = 0
        for i in range(1, n +1):
            x = a + i * width
            area_sum += f(x)

        x = np.linspace(a, b, n+1)
        y = f(x)
        plt.bar(x[1:], y[1:], width=-width, align="edge", edgecolor="black")
        plt.plot(x[1:], y[1:], marker="o", color="black", markersize = 10)
        plt.show();
        plt.clf()
        return area_sum * width
```

```python
def middle_sum(func, a, b, n, kind="equal"):
    def f(x):
        return eval(func)

    if kind == "equal":
        width = abs((b - a)/n)
        area_sum = 0
        for i in range(1, n +1):
            x = a + (i - 0.5) * width
            area_sum += f(x)

        x = np.linspace(a, b, n+1)
        y = [f(a + (i - 0.5) * width) for i in x]
        plt.bar(x[1:], y[1:], width=width, align="edge", edgecolor="black")
        plt.plot(x[1:], y[1:], marker="o", color="black", markersize = 10)
        plt.show();
        plt.clf()
        return area_sum * width
```

```python
result = [left_sum("4*x**2 + 2 * x - 1", 2, 6, x) for x in [4, 5, 10, 100, 100]]
print(result)
```

```python
result = [right_sum("4*x**2 + 2 * x - 1", 2, 6, x) for x in [4, 5, 10, 100, 100]]
print(result)
```

```python
result = [middle_sum("4*x**2 + 2 * x - 1", 2, 6, x) for x in [4, 5, 10, 100, 100]]
print(result)
```
